```python
carrys = wwc2023_events["carrys"]
passes = wwc2023_events["passes"]
```

```python
match_id = 3893791

aitana_match_carrys_mask = (carrys.player == "Aitana Bonmati Conca") & (carrys.match_id == match_id)
aitana_match_carrys = carrys[aitana_match_carrys_mask]
```

```python
aitana_match_passes_mask = (passes.player == "Aitana Bonmati Conca") & (passes.match_id == match_id)
aitana_match_passes = passes[aitana_match_passes_mask]
```

```python
aitana_match_passes.id
```

```python
# Create a function to get the related pass event IDs
def get_related_pass_ids(related_events):
    pass_ids = []
    for event_id in related_events:
        if event_id in aitana_match_passes.id:
            pass_ids.append(event_id)
    return pass_ids

# Apply the function to each row in the carry DataFrame
aitana_match_carrys['related_pass_event'] = aitana_match_carrys['related_events'].apply(get_related_pass_ids)
aitana_match_carrys
```
